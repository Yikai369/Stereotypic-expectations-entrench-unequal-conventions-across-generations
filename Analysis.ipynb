```python
#@title Generate a csv file containing all the action data (Study 1 and 2)

size_conditions = [30, 100, 300, 600]
study_number = 1 # 1 or 2
num_runs = 5 if study_number == 1 else 30
prop = 0.5

if study_number == 1:
  save_dir = cwd + '/save_extended/Study1/'
elif study_number == 2:
  save_dir = cwd + '/save_extended/Study2/'

df_target = pd.DataFrame(columns=['chop', 'mine', 'build', 'sell_wood', 'sell_stone',
                                  'buy_wood', 'buy_stone', 'is_minority','group_identity',
                                  'epoch','run', 'population_size'])

for population_size in size_conditions:
  for run in range(num_runs):
    for agent_group in agent_group_of_interest:
      if study_number == 1:
        dir_group_data = dir_study1_group_data
      elif study_number == 2:
        dir_group_data = dir_study2_group_data

      df = pd.read_csv(dir_group_data + f'extended_2000epoch_study{study_number}_{population_size}_{prop}_{run}_agent{agent_group+1}_actions.csv', header=None)

      is_minority = [1 if agent_group in [1,3] else 0 for _ in range(len(df))]

      group_identity = [0 if agent_group in [0,1] else 1 for _ in range(len(df))]
      total_n_actions = np.array(df.iloc[0]).sum()
      df = df/total_n_actions
      df.columns = ['chop', 'mine', 'build', 'sell_wood', 'sell_stone', 'buy_wood', 'buy_stone']

      df['is_minority'] = is_minority
      df['group_identity'] = group_identity
      df['epoch'] = [i for i in range(len(df))]
      df['run'] = [run for _ in range(len(df))]
      df['population_size'] = [population_size for _ in range(len(df))]

      df_target = pd.concat([df_target,df])

df_target.to_csv(save_dir + f'extended_2000epoch_study{study_number}_agent_actions.csv', index=False, sep=',')
# df_target
```
